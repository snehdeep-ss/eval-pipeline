```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
df = pd.read_csv('metrics.csv')
df.head()
```

```python
# Latency distribution
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for ax, col, label in zip(axes, ['total_time', 'ttft', 'tpot'], ['Total Latency (s)', 'TTFT (s)', 'TPOT (s/tok)']):
    data = df[col].dropna()
    ax.hist(data, bins=20, edgecolor='black')
    for p, ls in [(50, '--'), (95, '-.'), (99, ':')]:
        v = data.quantile(p/100)
        ax.axvline(v, linestyle=ls, label=f'P{p}={v:.3f}')
    ax.set_title(label)
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig('latency_distribution.png', dpi=120)
plt.show()
```

```python
# Latency vs prompt length
plt.figure(figsize=(7, 4))
plt.scatter(df['prompt_len'], df['total_time'], alpha=0.6)
plt.xlabel('Prompt length (words)')
plt.ylabel('Total latency (s)')
plt.title('Latency vs Prompt Length')
plt.tight_layout()
plt.savefig('latency_vs_prompt_len.png', dpi=120)
plt.show()
```

```python
# Summary table
summary = df[['total_time', 'ttft', 'tpot', 'tokens_generated']].describe(percentiles=[.5, .95, .99])
print(summary.to_string())
```
